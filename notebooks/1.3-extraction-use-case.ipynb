# Extracting Product Information

```python
# Extraction of customer feedback
from openai import OpenAI
from pydantic import BaseModel, Field

client = OpenAI()

class Product(BaseModel):
    name: str = Field(description="The name of the product")
    price: float = Field(description="The price of the product")
    description: str = Field(description="The description of the product if available in the raw data, if not available return 'N/A'")

def extract_product_info(product_raw_data: str) -> str:
    response = client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are an expert at extracting product information from raw data."},
            {"role": "user", "content": product_raw_data}
        ],
        response_format=Product
    )
    return response.choices[0].message.parsed
```

# Extracting structured data from customer tickets to automate customer support

```python
import pandas as pd

csv_filename = "./extracted_ticket_issues.csv"
df = pd.read_csv(csv_filename)
df.head()
```

```python
from pydantic import BaseModel
from typing import Literal

class IssueCategory(BaseModel):
    category: Literal[
        "Billing Issues",
        "Account Access Issues", 
        "Product Information Requests",
        "Order Delays",
        "Return Assistance"
    ] = Field(description="The category of the issue")

def extract_issue_categories(all_issues: str) -> list[IssueCategory]:
    response = client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[{"role": "system", "content": "You are an expert at extracting issue categories from customer issues."}, {"role": "user", "content": all_issues}],
        response_format=IssueCategory
    )
    return response.choices[0].message.parsed.category

issue_examples = df['issue_description'].tolist()
issue_example = issue_examples[0]
issue_categories = extract_issue_categories(issue_example)
print(issue_categories)
```

```python
# Extract categories for all issues
all_categories = []
for issue in df['issue_description']:
    category = extract_issue_categories(issue)
    # Take first category if multiple are returned
    all_categories.append(category)

# Add new column with categories
df['issue_category'] = all_categories

# Display updated dataframe
print("\nDataframe with categorized issues:")
df.head()
```

# Organizing your Files and Notes

```python
# List all files in ./example-notes-files
# Create pydantic/openai data structure extraction for categorizing the types of files
# Run the categorization for all files and save the results to a dataframe with the original file name and its type

import os

files = os.listdir('./example-notes-files')
print(files)
```

```python
from pypdf import PdfReader

class FileCategory(BaseModel):
    file_category: str = Field(description="The category of the file")


def extract_file_types(file_contents: str) -> FileCategory:
    response = client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[{"role": "system", "content": "You are an expert at categorizing files and notes."}, {"role": "user", "content": file_contents}],
        response_format=FileCategory
    )
    return response.choices[0].message.parsed

def load_pdf_text(file_path):
    '''Loads text from a PDF file.'''
    # creating a pdf reader object
    reader = PdfReader(file_path)

    # extracting text from page
    text = "\n\n".join([page.extract_text() for page in reader.pages])
    
    return text

def load_file(file_name: str) -> str:
    """Loads .txt, .md, .pdf files"""
    if file_name.endswith('.txt'):
        return open(file_name, 'r').read()
    elif file_name.endswith('.md'):
        return open(file_name, 'r').read()
    elif file_name.endswith('.pdf'):
        return load_pdf_text(file_name)
    

all_file_categories = {"category": [], "file_name": []}
for file in files:
    file_path = f'./example-notes-files/{file}'
    file_contents = load_file(file_path)
    file_category = extract_file_types(file_contents)
    all_file_categories["category"].append(file_category.file_category)
    all_file_categories["file_name"].append(file)

df = pd.DataFrame(all_file_categories)
df.head()
```
